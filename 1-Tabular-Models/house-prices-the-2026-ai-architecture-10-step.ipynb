This is a complex <b>Regression</b> challenge handled with modern Neural Networks.</p><p style="font-size: 15px; line-height: 1.7; color: #8892b0;"><b>Evaluation Metric:</b> RMSE (Root-Mean-Squared-Error) between the log of the predicted value and the log of the observed sales price.</p></div><p style="text-align: center; margin-top: 25px; font-size: 13px; color: #495670; font-style: italic;">"Transforming real estate data into predictive market intelligence."</p></div>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
import warnings
warnings.filterwarnings('ignore')
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

# STEP 1: Understanding the Problem (Regression)
# STEP 2: Data Loading & Exploratory Data Analysis (EDA)
train = pd.read_csv('/kaggle/input/competitions/house-prices-advanced-regression-techniques/train.csv')

print("--- Data Summary (Step 2) ---")
train.info() # you want to see this
display(train.describe().T) # you like this summary

# 2026 Style Visualization
plt.style.use('dark_background')
plt.figure(figsize=(10, 6))
sns.histplot(train['SalePrice'], kde=True, color='#64ffda')
plt.title('Sale Price Distribution | 2026 Perspective', color='#64ffda', fontsize=16)
plt.show()
```

```python
# STEP 3 & 4: Feature Selection & Handling Object Types
# STEP 5 & 6: Missing Values & Feature Engineering
train['TotalSF'] = train['TotalBsmtSF'] + train['1stFlrSF'] + train['2ndFlrSF']
train['TotalBath'] = train['FullBath'] + (0.5 * train['HalfBath']) + train['BsmtFullBath']

# STEP 7: Categorical Encoding (get_dummies)
features = ['TotalSF', 'OverallQual', 'GrLivArea', 'GarageCars', 'YearBuilt', 'TotalBath']
X = pd.get_dummies(train[features], drop_first=True)
X = X.fillna(X.median()) # Handle remaining NaNs

# STEP 8: Target Transformation & Scaling
y = np.log1p(train['SalePrice']) 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"Feature Engineering Complete. Matrix Shape: {X_scaled.shape}")
```

```python
# STEP 9: Build and Fit Neural Network
model = Sequential([
    Dense(128, activation='relu', input_shape=[X_scaled.shape[1]]),
    BatchNormalization(), 
    Dropout(0.3),         # No memorization allowed!
    
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),         
    
    Dense(1)              
])

model.compile(optimizer='adam', loss='mse')

# STEP 10: Performance Assessment (Smart Training)
early_stop = EarlyStopping(
    monitor='val_loss', 
    patience=15, 
    restore_best_weights=True 
)

history = model.fit(
    X_scaled, y, 
    validation_split=0.2, # Self-testing during training
    epochs=150, 
    batch_size=32, 
    callbacks=[early_stop], 
    verbose=1
)

# Final Result
val_loss = min(history.history['val_loss'])
print(f"✅ Final Validation Loss (MSE): {val_loss:.4f}")
```

```python
import joblib
# Save for Hugging Face
model.save('house_model.keras')
joblib.dump(scaler, 'house_scaler.pkl')
print("Model and Scaler exported for AI Deployment! 🚀")
```

```python
# --- KAGGLE COMPETITION SUBMISSION ---

# 1. Load the unseen test data (The Final Exam)
test = pd.read_csv('/kaggle/input/competitions/house-prices-advanced-regression-techniques/test.csv')
test_ids = test['Id'] # Save IDs for the final Kaggle format

# 2. Apply the exact same Feature Engineering (The Secret Sauce)
test['TotalSF'] = test['TotalBsmtSF'] + test['1stFlrSF'] + test['2ndFlrSF']
test['TotalBath'] = test['FullBath'] + (0.5 * test['HalfBath']) + test['BsmtFullBath']

# 3. Select same features and apply One-Hot Encoding
X_test_submit = pd.get_dummies(test[features], drop_first=True)
X_test_submit = X_test_submit.fillna(X_test_submit.median()) # Fill missing values

# 4. CRITICAL: Align test columns with training columns (fill missing columns with 0)
X_test_submit = X_test_submit.reindex(columns=X.columns, fill_value=0)

# 5. Scale data using our fitted StandardScaler (No data leakage!)
X_test_scaled = scaler.transform(X_test_submit)

# 6. PREDICTION TIME!
log_preds = model.predict(X_test_scaled)

# 7. REVERSE LOG TRANSFORMATION: Convert predictions back to actual dollar amounts (expm1)
final_prices = np.expm1(log_preds).flatten()

# 8. Format and export the final submission file
submission = pd.DataFrame({
    'Id': test_ids,
    'SalePrice': final_prices
})
submission.to_csv('submission.csv', index=False)

print("✅ Championship file (submission.csv) is ready for Kaggle submission!")
```
