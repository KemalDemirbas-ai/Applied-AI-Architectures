```python
# ==============================================================================
# . SILENT MODE & WARNING SUPPRESSION
# ==============================================================================
import os
import warnings
import logging

# Suppress standard Python warnings
warnings.filterwarnings('ignore')

# Set TensorFlow log level: '3' suppresses all logs except fatal errors
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
# ==============================================================================
# 1. IMPORTING LIBRARIES AND SETTINGS (Adaptation of Steps 1 & 2)
# ==============================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Aesthetic Settings (Visual feast for the Kaggle community)
sns.set_theme(style="darkgrid", palette="pastel")
plt.rcParams['figure.figsize'] = (10, 6)

# Data paths (Configured for the Kaggle environment)
BASE_DIR = '/kaggle/input/competitions/cassava-leaf-disease-classification'
TRAIN_IMG_DIR = os.path.join(BASE_DIR, 'train_images/')
train_df = pd.read_csv(os.path.join(BASE_DIR, 'train.csv'))

# Mapping disease names
import json
with open(os.path.join(BASE_DIR, 'label_num_to_disease_map.json')) as f:
    disease_map = json.load(f)
    
train_df['disease_name'] = train_df['label'].astype(str).map(disease_map)
train_df['label_str'] = train_df['label'].astype(str) # String format for the Generator

# ==============================================================================
# 2. EDA: UNDERSTANDING THE SOUL OF THE DATA (Steps 2 & 3)
# ==============================================================================
print("\n--- Dataset Overview ---")
print(train_df.info())

# 📊 PLOT 1: Target Class Distribution (Neon and Vibrant)
plt.figure(figsize=(12, 6))
ax = sns.countplot(y='disease_name', data=train_df, palette="magma", 
                   order=train_df['disease_name'].value_counts().index)
plt.title("Distribution of Disease Classes (Watch out for Class Imbalance!)", fontsize=16, fontweight='bold')
plt.xlabel("Number of Images", fontsize=12)
plt.ylabel("Disease Name", fontsize=12)
plt.tight_layout()
plt.show() # The Kaggle bot will love to make this the thumbnail!

# ==============================================================================
# 3. DATA PREPARATION AND MANIPULATION (Adaptation of Steps 4, 5, 6, 7, 8)
# ==============================================================================
# Instead of get_dummies in tabular data, we use ImageDataGenerator here.
# We rotate and shift images to prevent overfitting (machine memorization).

BATCH_SIZE = 32
IMG_SIZE = (224, 224) # Standard size for EfficientNetB0

datagen = ImageDataGenerator(
    validation_split=0.2, # X_train and X_test (validation) split (Step 8)
    rescale=1./255,       # Normalize pixels to 0-1 range
    rotation_range=40,    # Rotate image to break memorization
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

train_generator = datagen.flow_from_dataframe(
    dataframe=train_df,
    directory=TRAIN_IMG_DIR,
    x_col="image_id",
    y_col="label_str",
    subset="training",
    batch_size=BATCH_SIZE,
    seed=42,
    shuffle=True,
    class_mode="categorical",
    target_size=IMG_SIZE
)

valid_generator = datagen.flow_from_dataframe(
    dataframe=train_df,
    directory=TRAIN_IMG_DIR,
    x_col="image_id",
    y_col="label_str",
    subset="validation",
    batch_size=BATCH_SIZE,
    seed=42,
    shuffle=False,
    class_mode="categorical",
    target_size=IMG_SIZE
)

# 📊 PLOT 2: Sample Images After Data Augmentation
images, labels = next(train_generator)
plt.figure(figsize=(14, 6))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(images[i])
    plt.title(f"Class: {np.argmax(labels[i])}", fontsize=10)
    plt.axis('off')
plt.suptitle("Manipulated (Augmented) Training Data the Model Will See", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# ==============================================================================
# 4. ADVANCED MODEL TRAINING (Step 9) - FINE-TUNING FOR HIGHER ACCURACY
# ==============================================================================
from tensorflow.keras.callbacks import ReduceLROnPlateau
from sklearn.utils import class_weight

# 1. Load Base Model and Partially Unfreeze (Break the chains)
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Strategy: Keep early layers frozen (basic shapes), unfreeze the last 20 layers (fine details).
base_model.trainable = True
for layer in base_model.layers[:-20]: 
    layer.trainable = False

# Build Custom Classification Head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.4)(x) # Increased penalty (from 0.3 to 0.4) to strictly prevent overfitting!
predictions = Dense(5, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# 2. Compile with a very fine Learning Rate (1e-4) for delicate tuning
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 3. Smart Control Mechanisms (Callbacks)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_cassava_model_v2.h5', save_best_only=True, monitor='val_accuracy')

# Speed Limiter: If val_loss stops improving for 2 epochs, reduce learning rate by 50%!
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6, verbose=1)

# 4. Handling Class Imbalance (Dynamic Class Weights)
# Calculate severe mathematical penalties for misclassifying minority disease classes
class_weights_array = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)
train_class_weights = dict(enumerate(class_weights_array))

print("\n--- Advanced Fine-Tuning Pipeline Initialized ---")
# Epochs increased to 20. Ensure GPU Accelerator is enabled!
history = model.fit(
    train_generator,
    validation_data=valid_generator,
    epochs=20, 
    class_weight=train_class_weights, # Reinforcements for minority classes!
    callbacks=[early_stopping, model_checkpoint, reduce_lr],
    verbose=1
)

# ==============================================================================
# 5. MEASURING MODEL PERFORMANCE (Step 10)
# ==============================================================================
# 📊 PLOT 3: Model Training Performance (Accuracy & Loss Curves)
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(len(acc))

plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Train Accuracy', color='cyan', linewidth=3)
plt.plot(epochs_range, val_acc, label='Validation Accuracy', color='magenta', linewidth=3)
plt.legend(loc='lower right', fontsize=12)
plt.title('Accuracy Plot', fontsize=14, fontweight='bold')
plt.xlabel('Epochs')
plt.ylabel('Score')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Train Loss', color='cyan', linewidth=3)
plt.plot(epochs_range, val_loss, label='Validation Loss', color='magenta', linewidth=3)
plt.legend(loc='upper right', fontsize=12)
plt.title('Loss Plot', fontsize=14, fontweight='bold')
plt.xlabel('Epochs')
plt.ylabel('Loss Value')

plt.suptitle("🚀 Model Training Journey (No Memorization, Steady Learning)", fontsize=18, fontweight='bold')
plt.show()

print("\n🏆 Process Complete! Model saved as 'best_cassava_model.h5'.")
# As the instructor requested, you can download this and deploy it live via Streamlit and Hugging Face!
```

```python
# ==============================================================================
# 6. KAGGLE SUBMISSION (GENERATING THE ANSWER KEY)
# ==============================================================================
import os
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing import image

print("\n--- Generating Kaggle Submission File ---")

# Path to the test images directory in Kaggle
TEST_DIR = '/kaggle/input/competitions/cassava-leaf-disease-classification/test_images'

predictions = []

# Iterate through each image in the test directory
for image_id in os.listdir(TEST_DIR):
    img_path = os.path.join(TEST_DIR, image_id)
    
    # Load and resize the image to the 224x224 format our model expects
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) # Add batch dimension
    img_array = img_array / 255.0 # Normalize pixels just like in training!
    
    # Let the AI make a prediction
    pred = model.predict(img_array, verbose=0) # verbose=0 keeps the console clean
    predicted_class = np.argmax(pred, axis=1)[0]
    
    # Append the result in the exact format Kaggle demands
    predictions.append({'image_id': image_id, 'label': predicted_class})

# Convert predictions to a Pandas DataFrame and save as 'submission.csv'
submission_df = pd.DataFrame(predictions)
submission_df.to_csv('submission.csv', index=False)

print("\n🏆 Process Complete! 'submission.csv' has been successfully created.")
print(submission_df.head())
```
