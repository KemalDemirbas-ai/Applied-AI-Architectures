    <h3 style="color:#00ffcc; margin-top:0;">🎯 Step 1: Project Definition & Metric</h3>
    <p style="color:#b0bec5; line-height:1.6;"><b>The Goal:</b> Predict unit sales for thousands of items sold at different Corporación Favorita grocery stores. This is a complex <b>Time Series Forecasting</b> challenge handled with modern Neural Networks.</p>
    <p style="color:#b0bec5; line-height:1.6;"><b>Evaluation Metric:</b> RMSLE (Root-Mean-Squared-Logarithmic-Error) between the predicted future sales and the actual sales.</p>

```python
# ==========================================
# ULTIMATE EDITION: STORE SALES ORACLE
# End-to-End Pipeline: EDA + Deep LSTM V2 + MLOps
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

# Theme Settings
plt.style.use('dark_background')
sns.set_palette("husl")

print("Phase 1: Data Detective (EDA) & Market Analysis...\n" + "="*55)

# 1. Loading Both Datasets
print("-> Loading Train and Test data...")
train_df = pd.read_csv('/kaggle/input/competitions/store-sales-time-series-forecasting/train.csv')
test_df = pd.read_csv('/kaggle/input/competitions/store-sales-time-series-forecasting/test.csv')
submission_ids = test_df['id']

# 2. Basic Data Inspection
print("\n--- Train Data General Info ---")
train_df.info()
print("\n--- Missing Values (NaN) Check ---")
print(train_df.isnull().sum())

# 3. EDA VISUALIZATION: MACRO TREND BLUEPRINT
print("\n-> Generating Macro Trend Blueprint...")
train_df['date'] = pd.to_datetime(train_df['date']) # Convert to datetime for plotting
daily_sales = train_df.groupby('date')['sales'].sum().reset_index()

plt.figure(figsize=(16, 6))
sns.lineplot(data=daily_sales, x='date', y='sales', color='#00ffcc', linewidth=1.5)
plt.title('Ecuador Store Sales: Macro Trend Blueprint (2013-2017)', color='#ff00ff', fontsize=18, fontweight='bold', pad=15)
plt.xlabel('Timeline', color='#b0bec5', fontsize=12)
plt.ylabel('Total Unit Sales', color='#b0bec5', fontsize=12)
plt.grid(color='#263238', linestyle='--', linewidth=0.5)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()

# ==========================================
print("\nPhase 2: Initiating Advanced Training Protocol (V2)...\n" + "="*55)

# 4. LOGARITHMIC TRANSFORMATION (Target Optimization)
print("-> Applying Logarithmic Transformation to Target Variable...")
y = np.log1p(train_df['sales']) 

# 5. Feature Engineering (Aligning Train & Test)
print("-> Merging datasets for aligned Feature Engineering...")
X = train_df.drop(['id', 'sales'], axis=1)
test_features = test_df.drop(['id'], axis=1)

X['is_test'] = 0
test_features['is_test'] = 1
combined = pd.concat([X, test_features])

# Time Extraction
combined['date'] = pd.to_datetime(combined['date'])
test_dates = combined[combined['is_test'] == 1]['date'] # Saving for final graph
combined['year'] = combined['date'].dt.year
combined['month'] = combined['date'].dt.month
combined['day'] = combined['date'].dt.day
combined['day_of_week'] = combined['date'].dt.dayofweek
combined = combined.drop(['date'], axis=1)

# One-Hot Encoding
print("-> Encoding Categories (Family & Stores)...")
combined = pd.get_dummies(combined, columns=['family', 'store_nbr'], drop_first=True)

# Splitting back to Train and Test
X_train_final = combined[combined['is_test'] == 0].drop('is_test', axis=1)
X_test_final = combined[combined['is_test'] == 1].drop('is_test', axis=1)

# Memory Optimization (CRITICAL)
print("-> Downcasting matrix to prevent RAM crash...")
for col in X_train_final.columns:
    X_train_final[col] = X_train_final[col].astype('float32')
    X_test_final[col] = X_test_final[col].astype('float32')

# Scaling and 3D Reshaping for LSTM
print("-> Scaling and Reshaping for Neural Engine...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_final)
X_test_scaled = scaler.transform(X_test_final)

X_train_lstm = np.expand_dims(X_train_scaled, axis=1)
X_test_lstm = np.expand_dims(X_test_scaled, axis=1)

# 6. Building the V2 Neural Engine
print("-> Compiling Deep LSTM Architecture...")
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2]), return_sequences=True))
model.add(LSTM(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dense(1)) 

model.compile(optimizer='adam', loss='mse')

# 7. Training Protocol
print("\n-> Commencing V2 Training Sequence...")
history = model.fit(X_train_lstm, y, epochs=5, batch_size=2048, validation_split=0.1, verbose=1)

# 8. Predicting & Reversing the Log Transformation
print("\n-> Predicting Future Sales and Reversing Log Transformation...")
log_predictions = model.predict(X_test_lstm)
actual_predictions = np.expm1(log_predictions.flatten()) 
actual_predictions = np.clip(actual_predictions, 0, None) 

# 9. Final Submission & Export
submission = pd.DataFrame({'id': submission_ids, 'sales': actual_predictions})
submission.to_csv('submission_v2.csv', index=False)
model.save('store_sales_lstm_v2.h5')

print("\n✅ MISSION ACCOMPLISHED: 'submission_v2.csv' and 'store_sales_lstm_v2.h5' generated!")

# ==========================================
# FINAL VISUALIZATION SHOW
# ==========================================
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12))

# Graph 1: Learning Curve
ax1.plot(history.history['loss'], label='Training Loss', color='#00ffcc', linewidth=2.5)
ax1.plot(history.history['val_loss'], label='Validation Loss', color='#ff00ff', linewidth=2.5)
ax1.set_title('Neural Engine V2 Learning Curve (Logarithmic Loss)', color='#ffffff', fontsize=16, pad=15)
ax1.set_xlabel('Epochs', color='#b0bec5')
ax1.set_ylabel('Loss', color='#b0bec5')
ax1.legend(facecolor='#111b2b', edgecolor='#263238', labelcolor='white')
ax1.grid(color='#263238', linestyle='--', linewidth=0.5)

# Graph 2: Future Prediction Forecast
ax2.plot(test_dates[:500], actual_predictions[:500], color='#00ffcc', linewidth=2)
ax2.set_title('V2 Future Sales Forecast (First 500 Predictions)', color='#ff00ff', fontsize=16, pad=15)
ax2.set_xlabel('Future Timeline', color='#b0bec5')
ax2.set_ylabel('Predicted Unit Sales', color='#b0bec5')
ax2.grid(color='#263238', linestyle='--', linewidth=0.5)

sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```
